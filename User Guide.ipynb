## 3. Load Data

Probably the first thing to do after importing your packages is to load your data. 
Hypy possesses a function to load text or xls file containing two columns of data without any heading or numbering. Below, you see an exemple of such a file : 
(Be careful to not run this cell, because it will probably send you an error as the file, data.png, will be missing)

```python
from IPython.display import Image
Image(filename = 'data.png')
```

If you have a file like that (the size of the columns has to be equal, but can be "infinite"), you can use the function
"load" from hypy to assign variables. The first column is the time and the second one is the drawdown. The variables chosen are often t and s, but this is up to you. To use "load", you have to type "hp.ldf('name of the file')". Be careful, your file must be in the same depository as your notebook. For the rest of the user guide, the file will be the data based on the "Theis" exemple ths_ds1.txt or ths_ds1.xlsx, both containing the same data. We will first start with the text file. 

```python
t,s = hp.ldftxt('ths_ds1.txt')
```

You can print your data to be sure that it loaded correctly

```python
print(t)
print(s)
```

You could have a file with multiple columns, or want to extract just one column. To do this, you can use numpy. 
the command is quite simple : t = np.loadtxt('ths_ds1.txt', usecols = 0). usecols is here to select the column you want. 0 is the first and so on. If you want the 4th one, usecols = 3. 

```python
t1 = np.loadtxt('ths_ds1.txt', usecols = 0)
s1 = np.loadtxt('ths_ds1.txt', usecols = 1)
```

To be sure, you can print the two arrays that should be the same. 

```python
print(t)
print(t1)
```

```python
print(s)
print(s1)
```

As you see, the result is the same. However, there is a difference that we can't see here. the function "ldftxt" transforms the data into a numpy array and deletes negative data that could disturb some of the calculations. To show this, a file was created with negative data within. The same operation will be done and the comparaison should be different now. 

```python
t1 = np.loadtxt('testsample.txt', usecols = 0)
s1 = np.loadtxt('testsample.txt', usecols = 1)
t,s = hp.ldftxt('testsample.txt')
```

```python
print(t)
print(t1)
```

```python
print(s)
print(s1)
```

```python
t1 = np.array(t1)
t1 = np.asfarray(t1,float)
condition = np.greater(t1,0)
t1 = np.extract(condition,t1)

s1 = np.array(s1)
s1 = np.asfarray(s1,float)
condition2 = np.greater(s1,0)
s1 = np.extract(condition2,s1)
```

```python
print(t)
print(t1)
```

```python
print(s)
print(s1)
```

This fixes the problem.

## 5. hypy

The rest of the functions you'll need are included in the hypy package. To use a specific function of a package, you need to seperate it with a "." . 

Hypy is divided in two parts : the module hypy, which contains all the generic functions and the specific modules such as "ths" or "pcw". In this chapter, it will show you how to use a function. The specific one chosen to do this is the Theis model "ths". If you're working with a different one, just change "ths" with the name of yours (for exemple "pcw").

If you want to use a function from the generic ones, just type hp.nameofthefunction(parameters). If you want one for the ths, type hp.ths.nameofthefunction(parameters). If you don't know the parameters, type "help nameofthefunction". After the dot (for exemple hp.) you can use tabulation. It will show you all the possible functions.

Starting now, this chapter assumes that you've already loaded the data and made a plot. But it will be shown again, quickly. 

For more details, you can look at the demo for each model, also available in Hypy. 

```python
t,s = hp.ldftxt('ths_ds1.txt')
```

```python
plt.plot(t,s)
plt.show()
```

Most of the time, you'll want to do a diagnostic plot :

```python
hp.diagnostic(t,s)
```

After that, you need to find a parameter p0. For that, you have to use the function guess (gss), which is specific to each model. Here, for theis ths, you have to type p0 = hp.ths.gss(t,s).

```python
p0 = hp.ths.gss(t,s)
```

you can check the quality of the guess with the function trial. if you're not sure about the parameters, type help(hp.trial).

```python
help(hp.trial)
```

```python
hp.trial(p0,t,s, name = 'ths')
```

The values are often good, but could be better. For this, you can call the function fit and check the new trial.

```python
p = hp.fit(p0,t,s,'ths')
hp.trial(p,t,s,'ths')
```

The last thing to do is to call the function repeat that sums up the work. For this, you usually need values that you mesured on the field, for exemple the pumping rate or the radial distance. The function rpt (repeat) contains many parameters. You can call help in order to adapt them if needed. For exemple, you can save it in pdf or png, add the name on the report, change the title, etc.

```python
q = 1.3888e-2 #pumping rate in m3/s
r = 250 #radial distance in m
d = [q,r]
```

```python
help(hp.ths.rpt)
```

```python
hp.ths.rpt(p,t,s,d,'ths',Author='Your name',ttle = 'Interference test', Rapport = 'My rapport', filetype = 'pdf')
```
